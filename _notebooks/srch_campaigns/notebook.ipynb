```python
import pandas as pd
```

```python
col = ["Campaign", "Ad Group", "Keyword", "Criterion Type"]
products = ["sofas", "convertible sofas", "love seats", "recliners", "sofa beds"]
words = ["buy", "prices", "bargain", "discount", "budget", "value"]
# kw = [f"{x} {y}" for x in products for y in words]+[f"{y} {x}" for x in products for y in words]
# print(f"{kw} length:{len(kw)}")

# products = [p * 6 for p in products]
# products *= 2
# print(products)

row01 = [["SEM_Sofas", p, f"{p} {w}", "Exact"] for p in products for w in words]
row02 = [["SEM_Sofas", p, f"{w} {p}", "Exact"] for p in products for w in words]
row = row01+row02
# print(row)
keywords_df = pd.DataFrame(data=row, columns=col)
keywords_csv = keywords_df.to_csv("keywords.csv")
print(keywords_csv)
```
